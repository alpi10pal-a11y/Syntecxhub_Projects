Compare NumPy performance with Python lists

```python
import time
```

```python
# python list
python_list = list(sepal_length)
start = time.time()
[x+1 for x in python_list]
end = time.time()
print("python list item :",end-start)
```

```python
start = time.time()
sepal_length + 1
end = time.time()
print("NumPy time:", end - start)
```
